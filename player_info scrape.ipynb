```python
import pandas as pd
import string
```

```python
#Example of what the table from each page looks like
url = 'https://www.basketball-reference.com/players/a/'
table = pd.io.html.read_html(url)
table[0]
```

```python
df = pd.DataFrame(columns=['Player', 'From', 'To', 'Pos', 'Ht', 'Wt', 'Birth Date', 'Colleges'])
for letter in list(string.ascii_lowercase): #
    url = 'https://www.basketball-reference.com/players/{}/'.format(letter)
    table = pd.io.html.read_html(url)[0]
    df = pd.concat([df, table])
```

```python
df.rename(columns={'Player':'player_name', 'Birth Date': 'birthDate'}, inplace=True)
```

```python
df.to_csv('player_info.csv', index=False)
```
